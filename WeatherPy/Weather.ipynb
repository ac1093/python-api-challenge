```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)


```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
city = "sydney"
params = {"q": city, "appid": weather_api_key, "units": "imperial"}

query_url = f"{base_url}appid={weather_api_key}&q={city}"


print(query_url)


response = requests.get(query_url, params= params)

city_weather = response.json()
City_name = city_weather['name']
Cloudiness = city_weather['clouds']['all']
Country = city_weather['sys']['country']
Date = city_weather['dt']
Humidity = city_weather['main']['humidity']
Lat = city_weather['coord']['lat']
Lng = city_weather['coord']['lon']
Max_temp = city_temp = city_weather['main']['temp_max']
Wind_speed = city_weather['wind']['speed']
City_ID = city_weather['id']
City_temp = city_weather['main']['temp']
print({f'the city name is {city} country is {Country} and the temperature is {City_temp}F lat{Lat},lng{Lng}'})
```

```python
weather_dict = ({"City":[],
                "Cloudiness":[],
                "Country":[],
                 "Date":[],
                 "Humidity":[],
                 "Lat":[],
                 "Lng":[],
                 "Max Temp":[],
                 "Wind Speed":[]})
```

```python


for city in cities:
    
    params = {"q": city, "appid": weather_api_key, "units": "imperial"}
    try:
        
        response = requests.get(base_url, params=params)
        city_weather = response.json()
        
        query_url = f"{base_url}appid={weather_api_key}&q={city}"
        
        time.sleep(0.5)
        
        City_name = city_weather['name']
        Cloudiness = city_weather['clouds']['all']
        Country = city_weather['sys']['country']
        Date = city_weather['dt']
        Humidity = city_temp = city_weather['main']['humidity']
        Lat = city_weather['coord']['lat']
        Lng = city_weather['coord']['lon']
        Max_temp = city_temp = city_weather['main']['temp_max']
        Wind_speed = city_weather['wind']['speed']
        City_ID = city_weather['id']
        City_temp = city_weather['main']['temp']
        
        if City_name not in weather_dict['City']:
            weather_dict['City'].append(City_name)
        
        if Cloudiness not in weather_dict['Cloudiness']:
            weather_dict['Cloudiness'].append(Cloudiness)
        
        if Country not in weather_dict['Country']:
            weather_dict['Country'].append(Country)
        
        if Date not in weather_dict['Date']:
            weather_dict['Date'].append(Date)
        
        if Humidity not in weather_dict['Humidity']:
            weather_dict['Humidity'].append(Humidity)
        
        if Lng not in weather_dict['Lng']:
            weather_dict['Lng'].append(Lng)
        
        if Lat not in weather_dict['Lat']:
            weather_dict['Lat'].append(Lat)
        
        if Max_temp not in weather_dict['Max Temp']:
            weather_dict['Max Temp'].append(Max_temp)
        
        if Wind_speed not in weather_dict['Wind Speed']:
            weather_dict['Wind Speed'].append(Wind_speed)
        
        
        print({f'The city name is {City_name} id {City_ID} and the temperature is {City_temp}F'})
        
        
    except:
        
        print("City not Found")
        pass

```
