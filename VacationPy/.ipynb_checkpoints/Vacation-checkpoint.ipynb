```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os


# Import API key
from api_keys import g_key
```

```python
cities_data = "../output_data/cities.csv"
cities_data = pd.read_csv(cities_data)
cities_data.head(10)
```

```python
best_weather = cities_data.loc[(cities_data['Max Temp']>70) & (cities_data['Cloudiness']<1) & (cities_data['Wind Speed']<10)]

best_weather_df = pd.DataFrame(best_weather)
best_weather_df
best_weather_df.to_csv(r'best_weather.csv')
```

```python
x_axis = cities_data['Lat']
y_axis = cities_data['Max Temp']
plt.title("City Latitude vs. Max Temperature 01/27/20")
plt.ylabel("MaxTemperature(F)")
plt.xlabel("Latitude")
plt.ylim(cities_data['Max Temp'].min()-10, cities_data['Max Temp'].max()+10, 5)
plt.xlim(cities_data['Lat'].min()-10, cities_data['Lat'].max()+10, 10)

plt.scatter(x_axis, y_axis, marker="o", color="red")
#plt.savefig("output_data/Maxt.png")
plt.show()
```

```python
Lat = cities_data["Lat"]
Lng = cities_data["Lng"]
locations = cities_data[["Lat", "Lng"]].astype(float) 
print( locations) 
```

```python
city_name = []


for index, row in cities_data.iterrows():
    if row["Cloudiness"] == 0:
        if row ["Wind Speed"] < 10:
            if row["Max Temp"] >70 and row["Max Temp"]<90:
                city_name.append(row)

city_weather_df = pd.DataFrame(city_name)

city_weather_df.to_csv("city_perfect_weather.csv")
city_weather_df.head()
```

```python
gmaps.configure(api_key=g_key)
fig = gmaps.figure()
locations = locations
humidity = cities_data["Humidity"]
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 2)
fig.add_layer(heat_layer)
#print(fig)
fig
```
